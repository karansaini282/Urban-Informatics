##### This notebook is set up to link as a default to the fb55 account.  To turn in the homework use the citibike dataset - check that you are able to read in about 40k rows

```python
API_URL = 'https://fb55.carto.com/api/v2/sql?q='

from __future__ import print_function

try:
    import urllib2 as ulib
except ImportError:
    from urllib.request import urlopen

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError

import ast
import pandas as pd

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))        
```

```python
test = '''
SELECT *
FROM citibike
'''
```

```python
get_data(test)
```

<h1 style='color:red'>TASK 1</h1>

 Sort data by start_station_id, tripduration
- Only checking trips with duration <= 3 hours

```python
test = '''
SELECT *
FROM citibike WHERE tripduration<=3*60*60 ORDER BY start_station_id, tripduration
'''
get_data(test)
```

Only show the top/last 10 records (aka head and tail in SQL)

```python
test = '''
SELECT *
FROM citibike ORDER BY cartodb_id LIMIT 10
'''
get_data(test)
```

```python
test = '''
SELECT *
FROM citibike ORDER BY cartodb_id DESC LIMIT 10
'''
get_data(test)
```

 List all unique start_station_id values

```python
test = '''
SELECT DISTINCT start_station_id
FROM citibike
'''
get_data(test)
```

 Aggregation functions:
- Count the number of trips (aka wc -l in SQL)
- Find the average/min/max trip duration

```python
test = '''
SELECT AVG(tripduration),MAX(tripduration),MIN(tripduration)
FROM citibike
'''
get_data(test)
```

<h1 style='color:red'>TASK 2</h1>

Selecting trips started on Feb-02-2015 only

```python
test = '''
SELECT *
FROM citibike WHERE EXTRACT(MONTH FROM starttime::date)=2 AND EXTRACT(YEAR FROM starttime::date)=2015 AND EXTRACT(DAY FROM starttime::date)=2
'''
get_data(test)
```

 Selecting trips started on the weekends
- What are average trip duration during weekends?

```python
test = '''
SELECT AVG(tripduration) FROM citibike WHERE EXTRACT(DOW FROM starttime::date) IN (0,6)
'''
get_data(test)
```

 Can we do the same for weekday?

```python
test = '''
SELECT AVG(tripduration) FROM citibike WHERE EXTRACT(DOW FROM starttime::date) NOT IN (0,6)
'''
get_data(test)
```

<h1 style='color:red'>TASK 3</h1>

Showing the list of start station locations
- Using GROUP BY


```python
test = '''
SELECT start_station_name
FROM citibike GROUP BY start_station_name
'''
get_data(test)
```

 Showing the number of trips started per station

```python
test = '''
SELECT start_station_name,count(*)
FROM citibike GROUP BY start_station_name
'''
get_data(test)
```

  but only for stations within 500m of Time Square!
- The coordinates of Time Square is (40.7577,-73.9857)

```python
#Distance in km = 0.5
test = '''
SELECT start_station_name,count(*)
FROM citibike WHERE start_station_name IN (SELECT DISTINCT start_station_name
FROM citibike WHERE 111.111*DEGREES(ACOS(COS(RADIANS(start_station_latitude)) * COS(RADIANS(40.7577)) * COS(RADIANS(start_station_longitude - -73.9857)) + SIN(RADIANS(start_station_latitude)) * SIN(RADIANS(40.7577)))) <=0.5) GROUP BY start_station_name
'''
get_data(test)
```

<h1 style='color:red'>TASK 4</h1>

 Find the station that had the longest average trip duration during
weekends and within 500m of TimeSquare! 

```python
test = '''
SELECT start_station_name,avg FROM(SELECT start_station_name,AVG(tripduration)
FROM citibike WHERE start_station_name IN (SELECT DISTINCT start_station_name
FROM citibike WHERE 111.111*DEGREES(ACOS(COS(RADIANS(start_station_latitude)) * COS(RADIANS(40.7577)) * COS(RADIANS(start_station_longitude - -73.9857)) + SIN(RADIANS(start_station_latitude)) * SIN(RADIANS(40.7577)))) <=0.5) AND EXTRACT(DOW FROM starttime::date) IN (0,6) GROUP BY start_station_name) AS avg_table1 WHERE avg IN (SELECT MAX(avg) FROM(SELECT start_station_name,AVG(tripduration)
FROM citibike WHERE start_station_name IN (SELECT DISTINCT start_station_name
FROM citibike WHERE 111.111*DEGREES(ACOS(COS(RADIANS(start_station_latitude)) * COS(RADIANS(40.7577)) * COS(RADIANS(start_station_longitude - -73.9857)) + SIN(RADIANS(start_station_latitude)) * SIN(RADIANS(40.7577)))) <=0.5) AND EXTRACT(DOW FROM starttime::date) IN (0,6) GROUP BY start_station_name) AS avg_table2)
'''
get_data(test)
```

Extra: create lines for trips started from stations within 500m of Times
Squares and lasted less than 2 hours. The number of trips per each
pair of stations are output as attributes of these lines.

```python
#Distance in km = 0.5
test = '''
SELECT start_station_name,end_station_name,count(*)
FROM citibike WHERE start_station_name IN (SELECT DISTINCT start_station_name
FROM citibike WHERE 111.111*DEGREES(ACOS(COS(RADIANS(start_station_latitude)) * COS(RADIANS(40.7577)) * COS(RADIANS(start_station_longitude - -73.9857)) + SIN(RADIANS(start_station_latitude)) * SIN(RADIANS(40.7577)))) <=0.5) AND tripduration<2*60*60 GROUP BY start_station_name,end_station_name
'''
get_data(test)
```

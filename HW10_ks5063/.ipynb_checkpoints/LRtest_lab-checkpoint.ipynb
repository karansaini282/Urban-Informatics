```python
import numpy as np
import scipy as sp
import statsmodels.formula.api as smf
import pylab as pl
import pandas as pd
%pylab inline
```

```python
df = pd.read_csv('https://github.com/fedhere/PUI2018_fb55/raw/master/Lab9_fb55/Asc.csv', header=None, names=['x', 'y1','tmp2', 'y2', 'tmp3', 'tmp4', 'tmp5', 'tmp6'], sep="\t")
```

```python
df
```

```python
df.sort_values(by='x', inplace=True)
```

```python
df
```

```python
pl.figure(figsize=(5,5))
pl.plot(df['x'], df['y1'], 'go')
pl.plot(df['x'], df['y2'], 'r*')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 1 & 2');
```

# First Anscomby distribution: y1

## OLS linear fit  y1~x

```python
OLS1 = smf.ols(formula='y1  ~ x', data=df).fit()
OLS1.summary()
```

```python
pl.figure(figsize=(5,5))
pl.plot(df['x'],df['y1'],'go')
pl.plot(df['x'], OLS1.predict(df),'-', color='orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 1 - line fit');
```

## OLS quadratic fit  y1~x^2 + x

```python
# create a new feature for x^2
df['x2'] = df['x']**2
```

```python
df.head()
```

```python
OLS2 = smf.ols(formula='y1 ~ x2 + x',data=df).fit()
OLS2.summary()
```

```python
df['y_pred'] = OLS2.predict()
df.sort_values(by="x",inplace=True)
pl.figure(figsize=(5,5))
pl.plot(df['x'],df['y1'],'go')
pl.plot(df['x'], df['y_pred'],'-', color='orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 1 - quadratic fit');
```

# Compare the models with the LR ratio test

```python
OLS2.compare_lr_test(OLS1)
```

# chi sq table for your convenience

![chi sq table](chisq.gif)

Equation of LR:
    
    LR = -2 log (Likelihood_restricted / likelihood_complex)
    
or equivalently

    LR = -2 (log(Likelihood)_restricted - log(Likelihood)_complex)

where log is the natural log (base e)

# Second Anscomby distribution: y2

## OLS linear fit  y2~x

```python
OLS2_1 = smf.ols(formula='y2  ~ x', data=df).fit()
OLS2_1.summary()
```

```python
pl.figure(figsize=(5,5))
pl.plot(df['x'],df['y2'],'go')
pl.plot(df['x'], OLS2_1.predict(df),'-', color='orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 2 - line fit');
```

## OLS quadratic fit  y2~x^2 + x

```python
OLS2_2 = smf.ols(formula='y2 ~ x2 + x',data=df).fit()
OLS2_2.summary()
```

```python
df['y_pred2'] = OLS2_2.predict()
pl.figure(figsize=(5,5))
pl.plot(df['x'],df['y2'],'go')
pl.plot(df['x'], df['y_pred2'],'-', color='orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Anscomby 2 - quadratic fit');
```

# Compare the models with the LR ratio test

```python
OLS2_2.compare_lr_test(OLS2_1)
```

<strong>
This is a Likelihood ratio test to test whether restricted model is correct. Over here OLS2_1 is the restricted model since it has a lesser degree of freedom.
<ul>
<li>146.63437980392047 - likelihood ratio, chisquare distributed with df_diff degrees of freedom (which is 1).
<li>9.4331834208378821e-34 - p-value of the test statistic. Hence, the test passes if alhpa = 0.05 and we can reject the Null hypothesis which is that "the simpler model is preferable".
<li>1.0 - degrees of freedom of the restriction, i.e. difference in df between models.
</ul>
</strong>

<strong><span style="color:red">Conclusion: </span>Second Anscomby distribution (y2) has much lower p-value than First Anscomby distribution (y1) which makes sense since y2 looks much more likely to be quadratic. It can  also be confirmed by the relative difference in R2 value (much larger for y2) for linear and quadratic models for each distribution.</strong>

# Calculating the LR test statistic manually

Equation of LR:
    
    LR = -2 log (Likelihood_restricted / likelihood_complex)
    
or equivalently

    LR = -2 (log(Likelihood)_restricted - log(Likelihood)_complex)

# First Anscomby distribution: y1

<strong>Calculating LR</strong>

```python
-2*(OLS1.llf - OLS2.llf)
```

<strong>Calculating P value</strong>

```python
1 - sp.stats.chi2.cdf(0.70794919019374447, 1)
```

# Second Anscomby distribution: y2

<strong>Calculating LR</strong>

```python
-2*(OLS2_1.llf - OLS2_2.llf)
```

<strong>Calculating P value</strong>

```python
1 - sp.stats.chi2.cdf(146.63437980392047, 1)
```
